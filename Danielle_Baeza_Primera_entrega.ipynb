# Análisis por Marcas



```python
marca_frecuencia = df['Brand'].value_counts()

# Calcular el rating promedio por marca
marca_rating_promedio = df.groupby('Brand')['Rating'].mean().sort_values(ascending=False)

# Calcular el número total de reseñas por marca
marca_reseñas = df.groupby('Brand')['Number_of_Reviews'].sum().sort_values(ascending=False)

# Mostrar resultados
print("Frecuencia de compra por marca:")
print(marca_frecuencia.head())

print("\nRating promedio por marca:")
print(marca_rating_promedio.head())

print("\nNúmero total de reseñas por marca:")
print(marca_reseñas.head())
```

```python
líderes_categoría = df.groupby(['Category', 'Brand']).agg({
    'Rating': 'mean',
    'Number_of_Reviews': 'sum',
    'Brand': 'count'  # Frecuencia
}).rename(columns={'Brand': 'Frecuencia'}).sort_values(by=['Frecuencia'], ascending=False).head(10) #Rename the column to 'Frecuencia' to avoid ambiguity
print(líderes_categoría)
```

```python
# Graficar las marcas más frecuentes
top_marcas_frecuencia = marca_frecuencia.head(10)
top_marcas_frecuencia.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Frecuencia de compra por Marca (Top 10)')
plt.xlabel('Marca')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# Graficar el rating promedio de las principales marcas
top_marcas_rating = marca_rating_promedio.head(10)
top_marcas_rating.plot(kind='bar', figsize=(10, 6), color='lightgreen')
plt.title('Rating promedio por Marca (Top 10)')
plt.xlabel('Marca')
plt.ylabel('Rating Promedio')
plt.xticks(rotation=45)
plt.show()

# Graficar las marcas con más reseñas
top_marcas_reseñas = marca_reseñas.head(10)
top_marcas_reseñas.plot(kind='bar', figsize=(10, 6), color='salmon')
plt.title('Número de reseñas por Marca (Top 10)')
plt.xlabel('Marca')
plt.ylabel('Total de Reseñas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Agrupar por categoría y calcular las métricas clave
categoría_marca = df.groupby(['Category', 'Brand']).agg({
    'Rating': 'mean',
    'Number_of_Reviews': 'sum',
    'Brand': 'count'  # Frecuencia de compra
}).rename(columns={'Brand': 'Frecuencia de compra'}).reset_index()

# Ordenar por frecuencia dentro de cada categoría
categoría_marca_líderes = categoría_marca.sort_values(
    by=['Category', 'Frecuencia de compra'], ascending=[True, False]
)

# Mostrar los líderes de cada categoría
print("Marcas líderes por categoría:")
print(categoría_marca_líderes.groupby('Category').head(1))
```

```python
import matplotlib.pyplot as plt

# Selecciona los 10 productos más reseñados
top_10_productos = df.groupby('Category').agg({
    'Rating': 'mean',
    'Number_of_Reviews': 'sum',
    'Usage_Frequency_Numeric': 'mean'  # Use the numeric column for Usage Frequency
}).sort_values(by='Number_of_Reviews', ascending=False).head(10)

top_10_productos.plot(kind='bar', figsize=(12, 6))
plt.title('Desempeño de los 10 productos más reseñados')
plt.ylabel('Valor')
plt.xticks(rotation=45)
plt.show()
```

Si analizamos el performance de los cosméticos el Serum, Face oil,  son los que tiene más reseñas, esto no significa que sea el mejor o peor ya que las reseñas nos hablan de cómo fue el producto desde el punto de vista del consumidor y es una tendencia significa al skin care que es lo cotidiano del día a día.



```python
# Promedio de rating por categoría
rating_por_categoria = df.groupby('Category')['Rating'].mean()

# Número total de reseñas por categoría
reseñas_por_categoria = df.groupby('Category')['Number_of_Reviews'].sum()

# Graficar comparación entre categorías
plt.figure(figsize=(12, 6))
rating_por_categoria.plot(kind='bar', color='yellow', label='Rating Promedio')
reseñas_por_categoria.plot(kind='bar', color='pink', alpha=0.6, label='Número de Reseñas')
plt.title('Comparación de Rating Promedio y Número de Reseñas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

Frecuencia de uso


1.   Correlacionar la frecuencia de uso con la categoría del producto o el tipo de piel.
2.   Comparar frecuencias entre diferentes grupos de usuarios








```python
# Calcular el promedio de 'Usage_Frequency' por categoría y tipo de piel
# Convert 'Usage_Frequency' to numerical representation if it's not already
frequency_mapping = {
    'Daily': 7,
    'Weekly': 1,
    'Monthly': 1/4,  # Assuming a month has 4 weeks
    'Occasional': 1/12  # Assuming occasional use is about once a month
    # Add more mappings if necessary
}
df['Usage_Frequency_Numeric'] = df['Usage_Frequency'].map(frequency_mapping)

# Now use the numeric column for the calculation
uso_promedio_categoria_piel = df.groupby(['Category', 'Skin_Type'])['Usage_Frequency_Numeric'].mean().unstack()

# Mostrar el resultado
print(uso_promedio_categoria_piel)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate average usage frequency by category and skin type
uso_promedio_categoria_piel = df.groupby(['Category', 'Skin_Type'])['Usage_Frequency_Numeric'].mean().unstack()

# Graficar un mapa de calor de la frecuencia de uso promedio
plt.figure(figsize=(12, 8))
sns.heatmap(uso_promedio_categoria_piel, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Frecuencia de Uso Promedio por Categoría de Producto y Skin_typt')
plt.xlabel('Skin_Type')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos ficticios de ejemplo
data = {
    'Precio': [25.5, 18.0, 30.0, 22.0, 27.5],
    'Tamaño': [200, 150, 300, 250, 180],
    'Número de Reseñas': [120, 80, 200, 150, 100],
    'Rating': [4.2, 3.8, 4.5, 4.0, 4.3]
}
df = pd.DataFrame(data)

# Calcular la matriz de correlación
correlation_matrix = df.corr()

# Mostrar la matriz de correlación
print("Matriz de correlación:")
print(correlation_matrix)

# Visualizar la matriz de correlación con un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```
